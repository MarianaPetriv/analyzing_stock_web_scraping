```python
import yfinance as yf
import pandas as pd
```

```python
apple = yf.Ticker("AAPL") #Using the Ticker module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is AAPL
```

```python
#!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json #Now we can access functions and variables to extract the type of data we need. You can view them and what they represent here https://aroussi.com/post/python-yahoo-finance.
```

```python
import json
with open('apple.json') as json_file:
    apple_info = json.load(json_file)
    # Print the type of data variable    
    #print("Type:", type(apple_info))
apple_info
```

```python
apple_info['country']
```

```python
apple_share_price_data = apple.history(period="max")
```

```python
apple_share_price_data.head()
```

```python
apple_share_price_data.reset_index(inplace=True) #We can reset the index of the DataFrame with the reset_index function. We also set the inplace paramter to True so the change takes place to the DataFrame itself.
```

```python
apple_share_price_data.plot(x="Date", y="Open") #We can plot the Open price against the Date:
```
